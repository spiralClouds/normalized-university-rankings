```python
import pandas as pd

data = pd.read_csv('data/raw/geo/worldcities.csv')
data.head()
```

```python

# convert city and country columns to lowercase
data['city'] = data['city'].str.lower()
data['country'] = data['country'].str.lower()

data_cleaned = data[['city', 'country', 'lat', 'lng']]
```

```python
#write to cleaned csv
data_cleaned.to_csv('data/cleaned/geo/worldcities.csv', index=False)
```
