```python
import pandas as pd

data = pd.read_csv('data/raw/geo/college_locations_covid.csv')

# make all column names lowercase
data.columns = [x.lower() for x in data.columns]
data.head()
```

```python
cleaned_data = data[['state', 'city', 'college']]

# make all city state and college names lowercase and trimed
cleaned_data['city'] = cleaned_data['city'].str.lower()
cleaned_data['state'] = cleaned_data['state'].str.lower()
cleaned_data['college'] = cleaned_data['college'].str.lower()
cleaned_data['country'] = 'united states'

cleaned_data.head()
```

```python
cleaned_data.to_csv('data/cleaned/geo/college_locations.csv', index=False)
```
